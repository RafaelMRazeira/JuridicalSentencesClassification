```python
import pandas as pd
```

```python
df = pd.read_csv('data/avaliacao_gabarito_490_v2.csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
labels = df['rotulo'].unique() 

percentages = df['rotulo'].value_counts()

fig1, ax1 = plt.subplots()

ax1.pie(percentages,
        labels=labels,
        autopct='%1.1f%%',
        shadow=True,
        startangle=90)
ax1.axis('equal') 
plt.show(), percentages
```

```python
regex_df = pd.read_json('data/Senteças_Classificadas.json',
                        orient='index')
```

```python
regex_df 
```

```python
labels = regex_df['classificacao'].unique() 

percentages = regex_df['classificacao'].str.lower().value_counts()

fig1, ax1 = plt.subplots()

ax1.pie(percentages,
        labels=labels,
        autopct='%1.1f%%',
        shadow=True,
        startangle=90)
ax1.axis('equal') 
plt.show(), percentages
```

```python
regex_df['classificacao'] = regex_df['classificacao'].apply(lambda x: x.lower()) 
regex_df
```
